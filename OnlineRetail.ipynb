```python
#Which region is generating the highest revenue?
highest_rev = df.groupby('Region')['Total  Revenue'].agg(sum)
highest_rev.sort_values(ascending=False).head(1) 
```

```python
#Which which region is generating the lowest?
lowest_rev = df.groupby('Region')['Total  Revenue'].agg(sum)
lowest_rev.sort_values(ascending=True).head(1) 
```

```python
Who are the top 10 customers and how much do they contribute to the total revenue?
top_customers = df.groupby('Customer Name')['Total  Revenue'].agg(sum).sort_values(ascending=False).head(10)
top_customers
```
