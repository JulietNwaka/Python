```python
# load first all date set and append
df1 = pd.read_excel('store.xlsx', sheet_name='Central')
df2 = pd.read_excel('store.xlsx', sheet_name='East')
df3 = pd.read_excel('south.xlsx', sheet_name='South')
df4 = pd.read_excel('store.xlsx', sheet_name='West')
```

```python
# append all table into a single table
df = pd.concat([df1,df2,df3,df4])
```

# Create a visulization for total revenue made in every region

```python

reg_sales = df.groupby('Region')['Total  Revenue'].agg(sum)
label = df['Region'].unique()

fig1, ax1 = plt.subplots()
ax1.pie(reg_sales,  labels=label, autopct='%1.1f%%')
ax1.axis('equal') 
plt.show()
```
